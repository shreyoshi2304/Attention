```python
import numpy as np
import matplotlib.pyplot as plt

# Stimulus space
stimulus = np.linspace(-5, 5, 200)

# Neurons
n_neurons = 30
preferred = np.linspace(-5, 5, n_neurons)

def tuning_curve(x, mu, sigma=1.0):
    return np.exp(-(x - mu)**2 / (2 * sigma**2))

# Baseline population response
responses = np.array([
    tuning_curve(stimulus, mu) for mu in preferred
])

# Attention profile (centered at target)
target = 1.0
attention = np.exp(-(preferred - target)**2 / 1.5)

# Apply attention
attended = responses * attention[:, None]

# Plot heatmap
plt.figure()
plt.imshow(attended, aspect='auto',
           extent=[stimulus.min(), stimulus.max(),
                   preferred.min(), preferred.max()])
plt.colorbar(label="Firing rate")
plt.xlabel("Stimulus")
plt.ylabel("Preferred stimulus")
plt.title("Population Coding with Attention")
plt.show()

# Compare summed activity
plt.figure()
plt.plot(stimulus, responses.sum(axis=0), label="Baseline")
print(" ")
plt.plot(stimulus, attended.sum(axis=0), label="With Attention")
plt.legend()
plt.title("Population Output")
plt.show()
```
