```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Time
T = 5
dt = 0.001
t = np.arange(0, T, dt)

# Signal
signal = np.sin(2 * np.pi * 6 * t)

# Time-varying attention
attention = 1 + 0.8 * np.sin(2 * np.pi * 0.5 * t)

# Noise
noise = np.random.normal(0, 0.4, len(t))

# Response
response = attention * signal + noise

# Plot
plt.figure()
plt.plot(t, response, label="Neural response")
plt.plot(t, attention, label="Attention signal", alpha=0.7)
plt.legend()
plt.title("Temporal Attention Modulation")
plt.xlabel("Time")
plt.show()
print("\n")

# ------------------------
# Temporal integration model
# ------------------------
tau = 0.05
V = np.zeros(len(t))

for i in range(len(t) - 1):
    input_signal = attention[i] * signal[i]
    dV = (-V[i] + input_signal) / tau
    V[i+1] = V[i] + dt * dV

plt.figure()
plt.plot(t, V)
plt.title("Integration under Attention")
plt.xlabel("Time")
plt.ylabel("State variable")
plt.show()
```
